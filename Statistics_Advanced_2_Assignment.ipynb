1. What is hypothesis testing in statistics?
- Hypothesis testing is a statistical method used to make inferences about a population based on sample data. It involves formulating two competing hypotheses: the null hypothesis (H0) and the alternative hypothesis (H1).The null hypothesis typically states that there is no effect or no difference in the population parameters, while the alternative hypothesis presents a different view, such as there being an effect or a difference. Through hypothesis testing, statistical techniques are used to determine whether there is enough evidence in the sample data to reject the null hypothesis in favor of the alternative hypothesis. This helps in drawing conclusions about the population based on the sample data.

2.  What is the null hypothesis, and how does it differ from the alternative
hypothesis?
- In hypothesis testing, the null hypothesis (H0) is a statement that there is no significant difference or relationship between variables or that any observed differences are due to chance. It is the default assumption that there is no effect or no difference in the population, and it is what we aim to either reject or fail to reject based on the data.
On the other hand, the alternative hypothesis (H1 or HA) is a statement that contradicts the null hypothesis. It suggests that there is a significant difference, effect, or relationship in the population, and it is what we aim to support if there is enough evidence.

3. Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.
- The significance level in hypothesis testing, often denoted by the symbol α (alpha), represents the probability of rejecting the null hypothesis when it is actually true. It measures how willing we are to make a Type I error, which is the error of rejecting a true null hypothesis.The significance level plays a crucial role in deciding the outcome of a hypothesis test. If the p-value is smaller than α, the null hypothesis is rejected, suggesting that there is a statistically significant difference or effect. On the other hand, if the p-value is greater than α, the null hypothesis is not rejected, indicating that there is not enough evidence to conclude that there is a significant difference.

4. What are Type I and Type II errors? Give examples of each.
- Type I error: This occurs when a true null hypothesis is incorrectly rejected. It is also known as a false positive. The probability of making a Type I error is represented by α (alpha), and is also known as the significance level of the test.
     - Example: A medical test incorrectly indicates that a person is pregnant when they are actually not.

- Type II error: This occurs when a false null hypothesis is failed to be rejected. It is also known as a false negative. The probability of making a Type II error is represented by β (beta).
    - Example: A medical test fails to detect a disease in a patient who actually has the disease.

5. What is the difference between a Z-test and a T-test? Explain when to use each.
- A Z-test and a T-test are both statistical hypothesis tests used to determine if there is a significant difference between the means of two groups. The main difference between the two tests lies in the assumptions made about the population variance.

1. Z-test:
- The Z-test is used when the population standard deviation is known.
- It is based on the standard normal distribution.
- Typically used when the sample size is large (usually n > 30).
- Provides a precise critical value for a given significance level.
- Often used in situations where the population standard deviation is known, such as in quality control or industrial applications.

2. T-test:
- The T-test is used when the population standard deviation is unknown and is estimated from the sample data.
- It is based on the t-distribution, which has heavier tails than the normal distribution.
- Typically used when the sample size is small (usually n < 30).
- Provides more conservative results compared to the Z-test, especially with small sample sizes.
- Often used in situations where the population standard deviation is unknown, such as in scientific research or medical studies.

```python
'''7. Implement hypothesis testing using Z-statistics for a sample dataset in
Python. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
                        50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
                        50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
                        50.3, 50.4, 50.0, 49.7, 50.5, 49.9] '''
import numpy as np
from scipy import stats
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

population_mean = 50.0
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)
sample_size = len(sample_data)
Z = (sample_mean - population_mean) / (sample_std / np.sqrt(sample_size))

p_value = 2 * (1 - stats.norm.cdf(abs(Z)))

print("Z-statistic:", Z)
print("p-value:", p_value)

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")

```

```python
 #8. Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

mean = 100
std_dev = 15
sample_size = 100

data = np.random.normal(mean, std_dev, sample_size)
confidence_interval = stats.norm.interval(0.95, loc=np.mean(data), scale=stats.sem(data))
plt.hist(data, bins=20, edgecolor='black')
plt.xlabel('Data')
plt.ylabel('Frequency')
plt.title('Histogram of Data from Normal Distribution')
plt.axvline(np.mean(data), color='r', linestyle='dashed', linewidth=1, label='Mean')
plt.axvline(confidence_interval[0], color='g', linestyle='dashed', linewidth=1, label='95% CI')
plt.axvline(confidence_interval[1], color='g', linestyle='dashed', linewidth=1)
plt.legend()
plt.show()
print("95% Confidence Interval for the Mean:", confidence_interval)
```

```python
#9.Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    z_scores = (data - np.mean(data)) / np.std(data)
    return z_scores

data = np.random.normal(0, 1, 1000)

z_scores = calculate_z_scores(data)
plt.hist(z_scores, bins=20, edgecolor='black')
plt.xlabel('Z-Scores')
plt.ylabel('Frequency')
plt.title('Histogram of Standardized Data (Z-Scores)')
plt.show()
```

1. What is the difference between a function and a method in Python?
- Function: A block of organized, reusable code that performs a specific task. It's called directly by its name.


```python
        def greet(name):  # Function
            return "Hello, " + name

        message = greet("Alice")
        print(message) # Output: Hello, Alice
```

- Method: A function that is associated with an object (an instance of a class). It operates on the data and properties of that object. Methods are called using the dot notation (object.method()).
  - Example:

```python
        class Dog:
            def __init__(self, name):
                self.name = name

            def bark(self):  # Method
                return "Woof!"

        my_dog = Dog("Buddy")
        sound = my_dog.bark()
        print(sound)  # Output: Woof!
```

3. What are the different ways to define and call a function in Python?

- Defining a Function:
  1. Standard Function Definition: Using the def keyword, followed by the function name, parameters (if any), and a colon. The function body is indented.


```python
        def add(x, y):
            return x + y
```

  2. Lambda Function (Anonymous Function): Using the lambda keyword to create a small, anonymous function. Lambda functions can only contain a single expression.


```python
        square = lambda x: x * x
```

  1.  Direct Call: Using the function name followed by parentheses, with the required arguments (if any).

```python
        result = add(5, 3)
        print(result)  # Output: 8
```

```python
        my_function = add
        result = my_function(2, 7)
        print(result)  # Output: 9
```

3. Calling a Lambda Function: Similar to calling a regular function, but using the variable to which the lambda function is assigned.

4. What is the purpose of the return statement in a Python function?
- The return statement serves two main purposes:
  1. Specify a value to be returned: The return statement sends a value back to the caller of the function. This value can be of any data type (e.g., integer, string, list, etc.). If no return statement is provided, the function implicitly returns None.

```python
    def multiply(x, y):
        result = x * y
        return result

    product = multiply(6, 7)
    print(product) # Output: 42
```

2. Terminate the function's execution: When a return statement is encountered, the function's execution immediately stops, even if there are more statements in the function body. Any code after the return statement will not be executed.

```python
    def example():
        print("Before return")
        return 10
        print("After return") # This line will NOT be executed

    value = example()
    print(value) # Output: 10
```

5. What are iterators in Python and how do they differ from iterables?
-  Iterable: An object that can be iterated over (i.e., you can loop through its elements). Examples of iterables include lists, tuples, strings, dictionaries, and sets. Iterables have the __iter__() method, which returns an iterator object.
  - Example:

```python
        my_list = [1, 2, 3]  # Iterable
        for item in my_list:
            print(item)
```

- Iterator: An object that produces a sequence of values one at a time. It has the __next__() method, which returns the next item in the sequence. When there are no more items, the __next__() method raises a StopIteration exception. Iterators are also iterables (they have an __iter__() method that returns themselves).
  -  Example:

```python
        my_list = [1, 2, 3]
        my_iterator = iter(my_list)  # Get an iterator from the list

        print(next(my_iterator))  # Output: 1
        print(next(my_iterator))  # Output: 2
        print(next(my_iterator))  # Output: 3
        #print(next(my_iterator))  # Raises StopIteration
```

- Key Differences:
1. An iterable is a container of data that can be iterated over.
2.  An iterator is an object that provides a way to access the elements of an iterable one at a time.
3.  An iterable can produce multiple iterators. An iterator is a "one-shot" object; once it's exhausted, it cannot be reused.

6. Explain the concept of generators in Python and how they are defined.
- Generator: A special type of function that returns an iterator object. Generators use the yield keyword to produce a sequence of values. Unlike regular functions, generators do not store the entire sequence in memory at once; instead, they generate values on demand.
- Definition: Generators are defined using the def keyword, just like regular functions, but they include at least one yield statement.

```python
        def my_generator(n):
            for i in range(n):
                yield i * 2

        # Using the generator
        for value in my_generator(5):
            print(value)
        # Output:
        # 0
        # 2
        # 4
        # 6
        # 8
```

7. What are the advantages of using generators over regular functions?
  1. Memory Efficiency: Generators generate values on demand, one at a time, rather than storing the entire sequence in memory. This is particularly beneficial when dealing with large datasets or infinite sequences.
  2. Improved Performance: Since generators don't create and store the entire sequence upfront, they can be more efficient in terms of processing time. They only compute the values that are actually needed.
  3. Simplified Code: Generators can make code more readable and concise, especially when dealing with complex iterative processes.
  4. Ability to represent infinite sequences: Because generators only calculate values when requested, they can easily represent infinite sequences that would be impossible to store in memory.


8. What is a lambda function in Python and when is it typically used?
- Lambda Function: An anonymous (unnamed) function defined using the lambda keyword. It can take any number of arguments but can only contain a single expression. The expression's result is automatically returned.
  - Syntax: lambda arguments: expression
- When to Use:
    -  Simple Operations: When you need a small, one-line function for a simple task.
    -  Higher-Order Functions: As arguments to functions like map(), filter(), sorted(), and reduce(), where you need to pass a function as an argument. These functions are known as higher-order functions.
    -  Short-Lived Functions: When you need a function that is only used once or in a limited scope.


```python
    # Example: Using lambda with map()
    numbers = [1, 2, 3, 4, 5]
    squared_numbers = list(map(lambda x: x**2, numbers))
    print(squared_numbers) # Output: [1, 4, 9, 16, 25]
```

9. Explain the purpose and usage of the map() function in Python.
- Purpose: The map() function applies a given function to each item in an iterable (e.g., list, tuple) and returns an iterator that yields the results.
  - Usage:
      1.   Syntax: map(function, iterable)
      2.  function: The function to apply to each item.
      3.  iterable: The iterable to process.

```python
    def double(x):
        return x * 2

    numbers = [1, 2, 3, 4]
    doubled_numbers = list(map(double, numbers))  # Use list() to get the results
    print(doubled_numbers)  # Output: [2, 4, 6, 8]

    # Using a lambda function with map()
    numbers = [5, 6, 7, 8]
    squared_numbers = list(map(lambda x: x * x, numbers))
    print(squared_numbers)  # Output: [25, 36, 49, 64]
```

10. What is the difference between map(), reduce(), and filter() functions in Python?
- These are all built-in functions used for functional programming in Python:
  -  map(function, iterable): Applies a given function to each item in an iterable and returns an iterator yielding the results. It transforms each element of the iterable.
  - filter(function, iterable): Filters elements from an iterable based on a given function. It returns an iterator containing only the elements for which the function returns True.


```python
    numbers = [1, 2, 3, 4, 5, 6]
    even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
    print(even_numbers) # Output: [2, 4, 6]
```

- reduce(function, iterable): Applies a function cumulatively to the items of an iterable, from left to right, so as to reduce the iterable to a single value. It is part of the functools module and needs to be imported.

```python
    from functools import reduce

    numbers = [1, 2, 3, 4]
    product = reduce(lambda x, y: x * y, numbers)
    print(product) # Output: 24 (1 × 2 × 3 * 4)
```

Key Differences:

•  map(): Transforms each element.
•  filter(): Selects elements based on a condition.
•  reduce(): Combines elements cumulatively to a single value.

```python
#1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.
import math
from functools import reduce

def sum_even_numbers(numbers):

    even_nums = [num for num in numbers if num % 2 == 0] # list comprehension, creates new list.
    return sum(even_nums) # returns the sum

num_list = [1, 2, 3, 4, 5, 6]
even_sum = sum_even_numbers(num_list)
print(f"The sum of even numbers in {num_list} is: {even_sum}") # Output: 12
```

```python
#4. Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(number):

    if number <= 1:
        return False
    if number <= 3:
        return True
    if number % 2 == 0 or number % 3 == 0:
        return False
    i = 5
    while i * i <= number:
        if number % i == 0 or number % (i + 2) == 0:
            return False
        i += 6
    return True

print("Prime numbers between 1 and 200:")

for num in range(1, 201):
    if is_prime(num):
        print(num, end=" ")
print()

```

```python
# 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
class FibonacciIterator:
    def __init__(self, limit):

        self.limit = limit
        self.a = 0
        self.b = 1
        self.count = 0

    def __iter__(self):
      return self

    def __next__(self):
        if self.count < self.limit:
            result = self.a
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return result
        else:
            raise StopIteration

fib_iter = FibonacciIterator(10) # creates object
print("Fibonacci sequence (10 terms):")
for num in fib_iter: # iterates and retrieves the data, calling the next method
    print(num, end=" ")
print()
```

```python
#6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def powers_of_2(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

power_gen = powers_of_2(5)
print("Powers of 2 (up to exponent 5):")
for power in power_gen:
    print(power, end=" ")
print()

```

```python
# 7. Implement a generator function that reads a file line by line and yields each line as a string.
def read_file_lines(filename):
    try:
        with open(filename, 'r') as file:
            for line in file:
                yield line.strip()  # Remove leading/trailing whitespace
    except FileNotFoundError:
        yield f"File '{filename}' not found." # handling error and returning a string

with open("my_file.txt", "w") as f:
    f.write("This is line 1.\n")
    f.write("This is line 2.\n")
    f.write("This is line 3.\n")

line_gen = read_file_lines("my_file.txt")
print("Contents of my_file.txt:")
for line in line_gen:
    print(line)
```

```python
# 10. Create a Python program that uses filter() to remove all the vowels from a given string
def remove_vowels(text):
    vowels = "aeiouAEIOU"
    return "".join(filter(lambda char: char not in vowels, text))

my_string = "Hello, World!"

consonants_only = remove_vowels(my_string)

print(f"Original string: {my_string}")
print(f"String without vowels: {consonants_only}")  # Output: Hll, Wrld!
```
